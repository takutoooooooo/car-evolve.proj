```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib
import matplotlib.ticker as ticker
```

```python
input_data_elite4 = pd.read_csv("../Assets/data/test_021511.csv")
input_data_elite8 = pd.read_csv("../Assets/data/test_021521.csv")
input_data_elite16 = pd.read_csv("../Assets/data/test_021532.csv")
input_data_elite2 = pd.read_csv("../Assets/data/test_021549.csv")
input_data_elite4_2 = pd.read_csv("../Assets/data/test_021606.csv")
input_data_elite1 = pd.read_csv("../Assets/data/test_021623.csv")
input_data_elite4.head()
```

```python
plt.plot(input_data_elite4["generations"], input_data_elite4[" 最高距離"], label = "エリート4")
plt.plot(input_data_elite8["generations"], input_data_elite8[" 最高距離"], label = "エリート8")
plt.plot(input_data_elite16["generations"], input_data_elite16[" 最高距離"], label = "エリート16")
plt.plot(input_data_elite2["generations"], input_data_elite2[" 最高距離"], label = "エリート2")
plt.plot(input_data_elite4_2["generations"], input_data_elite4_2[" 最高距離"], label = "エリート4_2")
plt.plot(input_data_elite1["generations"], input_data_elite1[" 最高距離"], label = "エリート1")
plt.title("世代数と最大距離の関係")
plt.xlabel("世代数")
plt.ylabel("最大到達距離")
plt.legend()
plt.gca().get_xaxis().set_major_locator(ticker.MaxNLocator(integer=True))
```

```python
plt.plot(input_data_elite4["generations"], input_data_elite4[" 平均"], label = "エリート4")
plt.plot(input_data_elite8["generations"], input_data_elite8[" 平均"], label = "エリート8")
plt.plot(input_data_elite16["generations"], input_data_elite16[" 平均"], label = "エリート16")
plt.plot(input_data_elite2["generations"], input_data_elite2[" 平均"], label = "エリート2")
plt.plot(input_data_elite4_2["generations"], input_data_elite4_2[" 平均"], label = "エリート4_2")
plt.plot(input_data_elite1["generations"], input_data_elite1[" 平均"], label = "エリート1")
plt.title("世代数と平均距離の関係")
plt.xlabel("世代数")
plt.ylabel("平均到達距離")
plt.legend()
plt.gca().get_xaxis().set_major_locator(ticker.MaxNLocator(integer=True))
```
