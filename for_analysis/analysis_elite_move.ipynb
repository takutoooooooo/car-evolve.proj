```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib
import matplotlib.ticker as ticker
```

```python
path = "../Assets/data/"
input_data_elite_move_8 = pd.read_csv(path + "elite_move/test_091607.csv")
input_data_elite8 = pd.read_csv(path + "ground_change/test_091440.csv")
input_data_elite_move_16 = pd.read_csv(path + "elite_move/test_091625.csv")
input_data_elite16 = pd.read_csv(path + "ground_change/test_091516.csv")
```

```python
plt.plot(input_data_elite_move_8["generations"], input_data_elite_move_8[" 最高距離"], label = "動的エリート8")
plt.plot(input_data_elite8["generations"], input_data_elite8[" 最高距離"], label = "静的エリート8")
plt.title("世代数と最大距離の関係")
plt.xlabel("世代数")
plt.ylabel("最大到達距離")
plt.legend()
plt.gca().get_xaxis().set_major_locator(ticker.MaxNLocator(integer=True))
```

```python
plt.plot(input_data_elite_move_16["generations"], input_data_elite_move_16[" 最高距離"], label = "動的エリート16")
plt.plot(input_data_elite16["generations"], input_data_elite16[" 最高距離"], label = "静的エリート16")
plt.title("世代数と最大距離の関係")
plt.xlabel("世代数")
plt.ylabel("最大到達距離")
plt.legend()
plt.gca().get_xaxis().set_major_locator(ticker.MaxNLocator(integer=True))
```

```python
plt.plot(input_data_elite_move_8["generations"], input_data_elite_move_8[" 平均"], label = "動的エリート8")
plt.plot(input_data_elite8["generations"], input_data_elite8[" 平均"], label = "静的エリート8")
plt.title("世代数と平均距離の関係")
plt.xlabel("世代数")
plt.ylabel("最大到達距離")
plt.legend()
plt.gca().get_xaxis().set_major_locator(ticker.MaxNLocator(integer=True))
```
