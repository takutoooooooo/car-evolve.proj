```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib
import matplotlib.ticker as ticker
```

```python
path_b = "../Assets/data/blx_alpha/"
path = "../Assets/data/ground_change/"
input_data_b_elite4 = pd.read_csv(path_b + "test_141359.csv")
input_data_b_elite4 = input_data_b_elite4[:70]
input_data_elite4 = pd.read_csv(path + "test_131616.csv")
input_data_elite4 = input_data_elite4[:70]
input_data_elite4.head()
```

```python
input_data_b_elite4["5日間移動平均"] = input_data_b_elite4[" 最高距離"].rolling(5).mean().round(1)
input_data_elite4["5日間移動平均"] = input_data_elite4[" 最高距離"].rolling(5).mean().round(1)
```

```python
plt.plot(input_data_b_elite4["generations"], input_data_b_elite4[" 最高距離"], label = "blx-a交叉")
plt.plot(input_data_elite4["generations"], input_data_elite4[" 最高距離"], label = "一様交叉")
plt.title("世代数と最大距離の関係")
plt.xlabel("世代数")
plt.ylabel("最大到達距離")
plt.legend()
plt.gca().get_xaxis().set_major_locator(ticker.MaxNLocator(integer=True))
```

```python
plt.plot(input_data_b_elite4["generations"], input_data_b_elite4[" 平均"], label = "blx-a交叉")
plt.plot(input_data_elite4["generations"], input_data_elite4[" 平均"], label = "一様交叉")
plt.title("世代数と平均距離の関係")
plt.xlabel("世代数")
plt.ylabel("平均到達距離")
plt.legend()
plt.gca().get_xaxis().set_major_locator(ticker.MaxNLocator(integer=True))
```

```python
plt.plot(input_data_b_elite4["generations"], input_data_b_elite4["5日間移動平均"], label = "blx-a交叉")
plt.plot(input_data_elite4["generations"], input_data_elite4["5日間移動平均"], label = "一様交叉")
plt.title("世代数と最大距離の関係：移動平均5世代")
plt.xlabel("世代数")
plt.ylabel("最大到達距離")
plt.legend()
plt.gca().get_xaxis().set_major_locator(ticker.MaxNLocator(integer=True))
```
