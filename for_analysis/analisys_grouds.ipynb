```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib
import matplotlib.ticker as ticker
```

```python
path = "../Assets/data/ground_change/"
# input_data_elite1 = pd.read_csv(path + "test_091355.csv")
input_data_elite1 = pd.read_csv(path + "test_131537.csv")
input_data_elite1 = input_data_elite1[:70]
# input_data_elite2 = pd.read_csv(path + "test_091408.csv")
input_data_elite2 = pd.read_csv(path + "test_131600.csv")
# input_data_elite4 = pd.read_csv(path + "test_091424.csv")
input_data_elite4 = pd.read_csv(path + "test_131616.csv")
# input_data_elite8 = pd.read_csv(path + "test_091440.csv")
input_data_elite8 = pd.read_csv(path + "test_131632.csv")
input_data_elite8 = input_data_elite8[:70]
# input_data_elite16 = pd.read_csv(path + "test_091516.csv")
input_data_elite16 = pd.read_csv(path + "test_141322.csv")
input_data_elite16 = input_data_elite16[:70]
# input_data_elite1.head()
```

```python
input_data_elite1["3日間移動平均"]=input_data_elite1[" 最高距離"].rolling(5).mean().round(1)
input_data_elite2["3日間移動平均"]=input_data_elite2[" 最高距離"].rolling(5).mean().round(1)
input_data_elite4["3日間移動平均"]=input_data_elite4[" 最高距離"].rolling(5).mean().round(1)
input_data_elite8["3日間移動平均"]=input_data_elite8[" 最高距離"].rolling(5).mean().round(1)
input_data_elite16["3日間移動平均"]=input_data_elite16[" 最高距離"].rolling(5).mean().round(1)
```

```python
plt.plot(input_data_elite1["generations"], input_data_elite1[" 最高距離"], label = "エリート1")
plt.plot(input_data_elite2["generations"], input_data_elite2[" 最高距離"], label = "エリート2")
plt.plot(input_data_elite4["generations"], input_data_elite4[" 最高距離"], label = "エリート4")
plt.plot(input_data_elite8["generations"], input_data_elite8[" 最高距離"], label = "エリート8")
plt.plot(input_data_elite16["generations"], input_data_elite16[" 最高距離"], label = "エリート16")
plt.title("世代数と最大距離の関係")
plt.xlabel("世代数")
plt.ylabel("最大到達距離")
plt.legend()
plt.gca().get_xaxis().set_major_locator(ticker.MaxNLocator(integer=True))
```

```python
plt.plot(input_data_elite1["generations"], input_data_elite1[" 平均"], label = "エリート1")
plt.plot(input_data_elite2["generations"], input_data_elite2[" 平均"], label = "エリート2")
plt.plot(input_data_elite4["generations"], input_data_elite4[" 平均"], label = "エリート4")
plt.plot(input_data_elite8["generations"], input_data_elite8[" 平均"], label = "エリート8")
plt.plot(input_data_elite16["generations"], input_data_elite16[" 平均"], label = "エリート16")
plt.title("世代数と平均距離の関係")
plt.xlabel("世代数")
plt.ylabel("平均到達距離")
plt.legend()
plt.gca().get_xaxis().set_major_locator(ticker.MaxNLocator(integer=True))
```

```python
plt.plot(input_data_elite1["generations"], input_data_elite1["3日間移動平均"], label = "エリート1")
plt.plot(input_data_elite2["generations"], input_data_elite2["3日間移動平均"], label = "エリート2")
plt.plot(input_data_elite4["generations"], input_data_elite4["3日間移動平均"], label = "エリート4")
plt.plot(input_data_elite8["generations"], input_data_elite8["3日間移動平均"], label = "エリート8")
plt.plot(input_data_elite16["generations"], input_data_elite16["3日間移動平均"], label = "エリート16")
plt.title("世代数と最大距離の関係：移動平均5世代")
plt.xlabel("世代数")
plt.ylabel("最大到達距離")
plt.legend()
plt.gca().get_xaxis().set_major_locator(ticker.MaxNLocator(integer=True))
```
